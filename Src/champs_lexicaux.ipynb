# Recherche de champs lexicaux avec Spacy ainsi que Gensim


```python
import spacy 
import gensim
import os
import re
from collections import Counter
```

```python
nlp = spacy.load("fr_core_news_sm")
```

```python
with open('/Users/morganr/Desktop/NLP_Naturalist_Corpus/Data/Naturaliste/Textes_Nettoyés/Goncourt/1860_Charles Demailly__travail.txt', 'r') as f:
    contenu = f.read()
print(contenu [:500])  # Affiche les 500 premiers caractères pour vérifier le contenu


doc = nlp(contenu)
```

```python
pos_tags = [token.pos_ for token in doc if not token.is_punct and not token.is_space]
```

```python
stats = Counter(pos_tags)
total = len(pos_tags)
total_mots = len(pos_tags)

```

```python
print(f"Analyse stylistique de : {'1860_Charles Demailly__travail.txt'}", f"Analyse sur {total_mots} mots :", sep="\n")


for pos, count in stats.most_common():
    print(f"{pos}: {count} ({(count/total)*100:.2f}%)")
```

```python
patterns = []
for i in range(len(doc) - 1):
    if doc[i].pos_ == "ADJ" and doc[i+1].pos_ == "NOUN":
        patterns.append(f"{doc[i].text} {doc[i+1].text}")
            
print(f"\nExemples de patterns ADJ+NOUN trouvés : {patterns[:5]}")
```
